```python
import pandas as pd
# ^^^ pyforest auto-imports - don't write above this line
from github import Github

# using an access token
g = Github("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
users = g.search_users('Machine Learning location:Lagos')
```

```python
name = []
login = []
bio = []
company = []
blog = []
email = []
followers = []
following = []
ID = []
url = []
location = []

for user in users:
    name.append(user.name)
    login.append(user.login)
    bio.append(user.bio)
    company.append(user.company)
    blog.append(user.blog)
    email.append(user.email)
    followers.append(user.followers)
    following.append(user.following)
    ID.append(user.id)
    url.append(user.url)
    location.append(user.location)
    
```

```python
data_tuples = list(zip(name,login,bio,company,blog,email,followers,following,ID,url,location))
```

```python
df1 = pd.DataFrame(data_tuples, columns=['name','login','bio', 'company', 'blog', 'email', 'followers', 'following',
                                        'ID', 'url', 'location'])
```
